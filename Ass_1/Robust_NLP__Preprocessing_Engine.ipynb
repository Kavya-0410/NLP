```python
### 1. Difference between "Love" and "love" in NLP

  # In Natural Language Processing, "Love" and "love" are treated as different tokens if case normalization is not applied.
  # This leads to duplication in the vocabulary and increases model complexity.
  # To avoid this, text is usually converted to lowercase so that both "Love" and "love" are treated as the same word.


### 2. What happens if stopwords are not removed?

  # If stopwords (such as "is", "the", "and") are not removed:
    # The dataset becomes larger and noisier
    # It increases computational cost
    # It may reduce model performance since these words carry little meaningful information
    #However, removing stopwords is not always beneficial and depends on the task.


### 3. Two real-world scenarios where removing stopwords can be harmful

  # 1. Sentiment Analysis
    # Example: "I am not happy"
    # If stopwords are removed → "happy"
    # This completely changes the meaning and gives incorrect sentiment.

  # 2. Question Answering Systems
    # Example: "What is your name?"
    # If stopwords are removed → "name"
    # This loses the actual intent of the question.


### 4. Difference between Stemming and Lemmatization

  #Stemming:
    # Reduces words to their root form by cutting suffixes
    # May produce non-meaningful words
    # Faster but less accurate
    # Example: "running" → "run", "studies" → "studi"

  #Lemmatization:
    # Converts words to their base form using a dictionary
    # Produces meaningful words
    # Slower but more accurate
    # Example: "running" → "run", "better" → "good"
```

```python
def preprocess_text(text):

    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove emails
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Lowercase
    text = text.lower()

    # Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters & emojis
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short words (<=2), keep "no", "not"
    cleaned_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            cleaned_tokens.append(word)

    # Join back to sentence
    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence
```

```python
sentence = "I am not happy with this!!! 😡😡"
tokens, cleaned = preprocess_text(sentence)

print("Tokens:", tokens)
print("Cleaned Sentence:", cleaned)
```

```python
##  Task 4: Token Analytics
```

```python
def token_analysis(tokens):

    if len(tokens) == 0:
        return 0, 0, 0

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(word) for word in tokens) / total_tokens

    return total_tokens, unique_tokens, avg_length
```

```python
for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    total, unique, avg_len = token_analysis(tokens)

    print("Sentence:", sentence)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg_len, 2))
    print("-" * 60)
```

```python
# Task 5: Frequency Analysis
```

```python
from collections import Counter

all_tokens = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)
```

```python
counter = Counter(all_tokens)
```

```python
print("Top 10 Most Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python
print("Top 10 Most Frequent Words:")
for word, count in counter.most_common(10):
    print(f"{word}: {count}")

print("\nTop 5 Least Frequent Words:")
for word, count in counter.most_common()[-5:]:
    print(f"{word}: {count}")
```

```python
##  Task 6: Full NLP Pipeline
```

```python
def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)

print("All Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
print(result["clean_sentences"])
```
